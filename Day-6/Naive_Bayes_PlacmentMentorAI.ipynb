```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
df = pd.read_csv('/content/placementmentor_nb_dataset.csv')
print(df.head())
```

```python
X = df[[
    'Python_Skill',
    'ML_Skill',
    'Cloud_Skill',
    'Networking_Skill',
    'SQL_Skill'
]]
```

```python
y=df['Predicted_Role']
```

```python
X_train,X_test,y_train,y_test=train_test_split(
    X,
    y,
    test_size = 0.2,
    random_state = 10
)
```

```python
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
acc = model.score(X_test,y_test)
print("Model Accuracy:",acc)
```

```python
y_predicted= model.predict(X_test)
print("Predicted Roles:")
print(y_predicted)
```

```python
pre = model.predict(
    [[90,85,76,88,99]]
)
print("Predicted Role:")
print(pre)
```

# **Multimonial_Naive_Bayes:**

```python
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.feature_extraction.text import CountVectorizer

from sklearn.naive_bayes import MultinomialNB



df = pd.read_csv('/content/multinomial_dataset.csv')

print(df.head())


X = df['Resume_Text']

y = df['Role']



cv = CountVectorizer()

X_vectorized = cv.fit_transform(X)



X_train, X_test, y_train, y_test = train_test_split(
    X_vectorized,
    y,
    test_size=0.2,
    random_state=10
)



model = MultinomialNB()



model.fit(X_train, y_train)



accuracy = model.score(X_test, y_test)

print("\nModel Accuracy:")

print(accuracy)


y_predicted = model.predict(X_test)

print("\nPredicted Roles:")

print(y_predicted)



sample_resume = [
    "Python TensorFlow Deep Learning NLP"
]

sample_vector = cv.transform(sample_resume)

prediction = model.predict(sample_vector)

print("\nPredicted Career Role:")

print(prediction)
```
