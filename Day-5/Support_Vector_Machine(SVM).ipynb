```python
# ==========================================
# STEP 1 — IMPORT LIBRARIES
# ==========================================

import pandas as pd

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

from sklearn.metrics import confusion_matrix

import seaborn as sn


# ==========================================
# STEP 2 — LOAD DATASET
# ==========================================

df = pd.read_csv('/content/multiclass_role_prediction_dataset.csv')

print(df.head())


# ==========================================
# STEP 3 — CHECK COLUMNS
# ==========================================

print(df.columns)


# ==========================================
# STEP 4 — DEFINE INPUT FEATURES (X)
# ==========================================

X = df.drop(['Predicted_Role'], axis='columns')


# ==========================================
# STEP 5 — DEFINE TARGET VARIABLE (y)
# ==========================================

y = df['Predicted_Role']


# ==========================================
# STEP 6 — TRAIN TEST SPLIT
# ==========================================

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=10
)


# ==========================================
# STEP 7 — CHECK TRAIN & TEST SIZE
# ==========================================

print(len(X_train))

print(len(X_test))


# ==========================================
# STEP 8 — CREATE SVM MODEL
# ==========================================

model = SVC()


# ==========================================
# STEP 9 — TRAIN MODEL
# ==========================================

model.fit(X_train, y_train)


# ==========================================
# STEP 10 — CHECK MODEL ACCURACY
# ==========================================

accuracy = model.score(X_test, y_test)

print("\nModel Accuracy:")

print(accuracy)


# ==========================================
# STEP 11 — PREDICT TEST DATA
# ==========================================

y_predicted = model.predict(X_test)

print("\nPredicted Roles:")

print(y_predicted)


# ==========================================
# STEP 12 — CONFUSION MATRIX
# ==========================================

cm = confusion_matrix(y_test, y_predicted)

print("\nConfusion Matrix:")

print(cm)


# ==========================================
# STEP 13 — VISUALIZE CONFUSION MATRIX
# ==========================================

plt.figure(figsize=(10,7))

sn.heatmap(cm, annot=True, fmt='d')

plt.xlabel('Predicted')

plt.ylabel('Truth')

plt.title('SVM Confusion Matrix')

plt.show()


# ==========================================
# STEP 14 — PREDICT NEW STUDENT ROLE
# ==========================================

# INPUT FORMAT:
# [Python_Skill, ML_Skill, Networking_Skill, Cloud_Skill, SQL_Skill]

prediction = model.predict([[85, 92, 20, 30, 75]])

print("\nPredicted Role:")

print(prediction)


# ==========================================
# STEP 15 — ANOTHER PREDICTION
# ==========================================

prediction2 = model.predict([[40, 25, 90, 35, 50]])

print("\nPredicted Role:")

print(prediction2)
```
