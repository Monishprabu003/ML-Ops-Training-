```python
# Import Libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# Load Dataset
df = pd.read_csv('/content/multiple_linear_regression_dataset.csv')

print("First 5 Rows of Dataset:\n")
print(df.head())
```

```python
X = df[[
    'Aptitude_Score',
    'Coding_Skill',
    'Communication_Skill',
    'Resume_Score',
    'Mock_Interview_Score'
]]

y = df['Placement_Readiness_Score']
```

```python

model = LinearRegression()

model.fit(X, y)
new_students = [
    [80, 75, 70, 85, 90],
    [60, 65, 55, 70, 75],
    [95, 90, 92, 88, 96]
]

predictions = model.predict(new_students)
```

```python

print("\nPredicted Placement Readiness Scores:\n")

for i in range(len(new_students)):
    print(f"Student {i+1}: {predictions[i]:.2f}")

# Model Accuracy
accuracy = model.score(X, y)

print("\nModel Accuracy:")
print(accuracy)
```
