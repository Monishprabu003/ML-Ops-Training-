```python
import pandas as pd

# Load Dataset
df = pd.read_csv('/content/one_hot_encoding_dataset.csv')

# Display First 5 Rows
print(df.head())
```

```python
print(df.dtypes)
```

```python
encoded_df = pd.get_dummies(df)

print(encoded_df.head())
```

```python
encoded_df = pd.get_dummies(df, drop_first=True)

print(encoded_df.head())
```

```python
from sklearn.preprocessing import OneHotEncoder

# Select categorical columns
categorical_cols = df[['Department', 'Company_Type', 'Programming_Language']]

# Create Encoder
encoder = OneHotEncoder(sparse_output=False)

# Apply Encoding
encoded_data = encoder.fit_transform(categorical_cols)

# Display Encoded Output
print(encoded_data)
```

```python
encoded_columns = encoder.get_feature_names_out([
    'Department',
    'Company_Type',
    'Programming_Language'
])

print(encoded_columns)
```

```python
encoded_df = pd.DataFrame(encoded_data, columns=encoded_columns)

print(encoded_df.head())
```

```python
# Drop original categorical columns
numerical_df = df.drop([
    'Department',
    'Company_Type',
    'Programming_Language'
], axis=1)

# Combine both
final_df = pd.concat([numerical_df, encoded_df], axis=1)

print(final_df.head())
```
