```python
# Import Libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Load Dataset
df = pd.read_csv('/content/multiple_linear_regression_dataset.csv')

# Display First 5 Rows
print("First 5 Rows of Dataset:\n")
print(df.head())

# Dataset Shape
print("\nDataset Shape:")
print(df.shape)

# Dataset Columns
print("\nDataset Columns:")
print(df.columns)

# Define INPUT FEATURES (X)
X = df[[
    'Aptitude_Score',
    'Coding_Skill',
    'Communication_Skill',
    'Resume_Score',
    'Mock_Interview_Score'
]]

# Define OUTPUT/TARGET (y)
y = df['Placement_Readiness_Score']

# Create Multiple Linear Regression Model
model = LinearRegression()

# Train the Model
model.fit(X, y)

# Predict for New Students
new_students = pd.DataFrame({
    'Aptitude_Score': [80, 60, 95],
    'Coding_Skill': [75, 65, 90],
    'Communication_Skill': [70, 55, 92],
    'Resume_Score': [85, 70, 88],
    'Mock_Interview_Score': [90, 75, 96]
})

predictions = model.predict(new_students)

# Display Predictions
print("\nPredicted Placement Readiness Scores:\n")

for i in range(len(new_students)):
    print(f"Student {i+1}: {predictions[i]:.2f}")

# Model Accuracy
accuracy = model.score(X, y)

print("\nModel Accuracy:")
print(accuracy)
```

```python
# Import Pickle
import pickle

# Save Model Using Pickle
with open('placement_model_pickle', 'wb') as file:
    pickle.dump(model, file)
```

```python
# Load Saved Model
with open('placement_model_pickle', 'rb') as file:
    mp = pickle.load(file)
```
