```python
# ==========================================
# STEP 1 — IMPORT LIBRARIES
# ==========================================

import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import confusion_matrix

import matplotlib.pyplot as plt
import seaborn as sn


# ==========================================
# STEP 2 — LOAD DATASET
# ==========================================

df = pd.read_csv('/content/multiclass_role_prediction_dataset.csv')

print(df.head())


# ==========================================
# STEP 3 — INPUT FEATURES (X)
# ==========================================

X = df.drop('Predicted_Role', axis='columns')


# ==========================================
# STEP 4 — TARGET VARIABLE (y)
# ==========================================

y = df['Predicted_Role']


# ==========================================
# STEP 5 — TRAIN TEST SPLIT
# ==========================================

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=10
)


# ==========================================
# STEP 6 — CREATE RANDOM FOREST MODEL
# ==========================================

# n_estimators = number of decision trees

model = RandomForestClassifier(
    n_estimators=20
)


# ==========================================
# STEP 7 — TRAIN MODEL
# ==========================================

model.fit(X_train, y_train)


# ==========================================
# STEP 8 — CHECK ACCURACY
# ==========================================

accuracy = model.score(X_test, y_test)

print("\nModel Accuracy:")

print(accuracy)


# ==========================================
# STEP 9 — PREDICT TEST DATA
# ==========================================

y_predicted = model.predict(X_test)

print("\nPredicted Roles:")

print(y_predicted)


# ==========================================
# STEP 10 — CONFUSION MATRIX
# ==========================================

cm = confusion_matrix(y_test, y_predicted)

print("\nConfusion Matrix:")

print(cm)


# ==========================================
# STEP 11 — VISUALIZE CONFUSION MATRIX
# ==========================================

plt.figure(figsize=(10,7))

sn.heatmap(cm, annot=True, fmt='d')

plt.xlabel('Predicted')

plt.ylabel('Truth')

plt.title('Random Forest Confusion Matrix')

plt.show()


# ==========================================
# STEP 12 — PREDICT NEW STUDENT ROLE
# ==========================================

# INPUT FORMAT:
# [Python_Skill, ML_Skill, Networking_Skill, Cloud_Skill, SQL_Skill]

prediction = model.predict([[85, 92, 20, 30, 75]])

print("\nPredicted Role:")

print(prediction)


# ==========================================
# STEP 13 — ANOTHER PREDICTION
# ==========================================

prediction2 = model.predict([[40, 25, 90, 35, 50]])

print("\nPredicted Role:")

print(prediction2)
```
